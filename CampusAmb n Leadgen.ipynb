```python
from selenium import webdriver
import time
import urllib.parse

driver = webdriver.Chrome()
driver.get("https://web.whatsapp.com")

input("👉 Scan QR and press ENTER...")

 #===== YOUR MESSAGE =====
message = """ Hello !!

We are hiring Campus Ambassador !!

Stipend up to 5000/-

Open for all students, who can represent us while working with our business team.

NO FEES FOR THIS.

Your opportunity involves :
🎯Letter of Recommendation
🎯After circulation you will receive official offer letter 
🎯Free Verified Certificates 
🎯Pre-Placement offer
Interested students can fill the details given below : 
NAME: 
EMAIL: 
COLLEGE: 
BRANCH/YEAR:
YOUR CONTACT  NUMBER:
Class representative number:
Your task is that !! 

Tomorrow we are going to organise one Updated industrial training and placement assistance program for the students of your college  

I will share you the invite link to you which we have created to share the detailed and further information.

You should add 80+ participants in the group by tomorrow.

Then I will send you offer letter to your mail within 24 hours.
All you need to do is just circulate the group link in your college groups.

Our team will organise the program tomorrow and I will let you know about the registrations.

Based up on that you wil get your stipend up to 10k
If you add 80+ members in group by tomorrow morning you will get campus ambassador certificate and based upon students registration you will get Stipends up to 5000
If you are interested then let me know.
Please find the requisite group link here: https://chat.whatsapp.com/F6saG8qh7WUBOv5jKIMpuj"""

while True:
    number = input("\nEnter number (or 'exit'): ")

    if number.lower() == "exit":
        break

    # Clean number
    number = number.replace("+", "").replace(" ", "")
    if not number.startswith("91"):
        number = "91" + number

    # Encode message
    encoded_msg = urllib.parse.quote(message)

    # Open chat with message pre-filled
    driver.get(f"https://web.whatsapp.com/send?phone={number}&text={encoded_msg}")

    print("⏳ Loading chat with message...")

    time.sleep(10)
    print(f"✅ Message loaded for {number}")
    print("👉 Edit if needed and press SEND manually")
    input("Press ENTER after sending...")

print("Done.")
```
